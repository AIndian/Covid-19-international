```python
# Dependencies
import pandas as pd
```

```python
file = 'response_measures.csv'
df_response = pd.read_csv(file)
df_response
```

```python
df_response = df_response.dropna(how='any')
```

```python
country = ['Belgium','Finland','France','Ghana','Egypt','Israel','Iraq','Vietnam','Singapore','Malaysia','United_Kingdom','United_States_of_America']
final_response = df_response.loc[df_response['Country'].isin(country)]
final_response
```

```python
sorted_response = final_response.sort_values(
    ["Country", "date_start"], ascending=True)

sorted_response['date_start'] = pd.to_datetime(sorted_response['date_start']).dt.strftime('%m/%d/%Y')
sorted_response['date_end'] = pd.to_datetime(sorted_response['date_end']).dt.strftime('%m/%d/%Y')
sorted_response
```

```python
sorted_response.to_csv("sorted_response.csv", index=False, header=True)
```

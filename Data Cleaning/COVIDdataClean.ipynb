```python
import pandas as pd
from datetime import datetime as dt
import os
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
import matplotlib.dates as mdates
import datetime as dt
import numpy as np
file = 'data.csv'
df = pd.read_csv(file)
df
```

```python
del df['countryterritoryCode']
del df['geoId']
df['date'] = pd.to_datetime(df[['month', 'day', 'year']])
del df['year']
del df['day']
df = df.loc[df['month'] != 1]
del df['month']
df = df.dropna(how='any')
del df['dateRep']

dfclean = df.loc[df['cases'] != 0]
dfclean = dfclean.rename(columns = {'dateRep':'Date'})
```

```python

df
```

```python
country = ['Belgium','Finland','France','Ghana','Egypt','Israel','Iraq','Vietnam','Singapore','Malaysia','United_States_of_America', 'United_Kingdom','Australia','New_Zealand']
final = dfclean.loc[dfclean['countriesAndTerritories'].isin(country)]
final = final.rename(columns = {'countriesAndTerritories': 'Country', 'continentExp': 'continent', 'Cumulative_number_for_14_days_of_COVID-19_cases_per_100000': '14 day cases/100,000', 'popData2019':'population'})
```

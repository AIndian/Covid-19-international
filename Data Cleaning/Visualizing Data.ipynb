```python
#Import dependencies
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stat
from scipy.stats import linregress
import matplotlib.dates as mdates
import matplotlib.cbook as cbook
```

```python
#Reading our cleaned data set
df = pd.read_csv("cleaned_data.csv")
df.head()
```

```python
#Looking for the unique array of sample continents
df["continent"].unique()
```

```python
#Setting each continent to its own dataframe
europe_df = df[df['continent'] == 'Europe']
africa_df = df[df['continent'] == 'Africa']
asia_df = df[df['continent'] == 'Asia']
america_df = df[df['continent'] == 'America']
```

```python
#Finding value counts for each sample country in Europe
europe_df["Country"].value_counts()
```

```python
#setting x-axis to date whenever country == United_Kingdom
#set y-axis to cases whenver country == United_kingdom
x_axisUK = europe_df[europe_df['Country'] == 'United_Kingdom']['Date']
y_axisUK = europe_df[europe_df['Country'] == 'United_Kingdom']['cases']

#Defining the parameters to calculate the R-Value
#(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axisUK, y_axisUK)
#regress_values = x_values * slope + intercept
#line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
#plt.scatter(x_axis,y_axis)
#plt.plot(x_axis,regress_values,"r-")
#plt.annotate(line_eq,(6,10),fontsize=15,color="red")

tsUK = pd.to_datetime(x_axisUK).dt.strftime('%m/%d/%Y')
fig, ax = plt.subplots()
plt.scatter(x_axisUK, y_axisUK, s=1, marker="o", color="red")
plt.title('UK COVID Rates Over Time')
plt.xlabel('Date')
plt.ylabel('Cases')

# Define the date format
yearsUK = mdates.YearLocator()   # every year
monthsUK = mdates.MonthLocator()  # every month
ax.xaxis.set_major_locator(monthsUK)

# Ensure a major tick for each Year using (interval=1)
#ax.xaxis.set_major_locator(years)
# ax.xaxis.set_major_formatter(years_fmt)
#ax.xaxis.set_minor_locator(months)
#ax.set_xlim(max(x_axis),min(x_axis))
#datelen = len(europe_df['Date'])
#stat.stats.linregress(europe_df[[x_axis, y_axis]].to_numpy()).rvalue ** 2
#ax.annotate("r-squared = {:.3f}".format(rvalue(x_axis, y_axis)), (0, 1))
ax.set_xlim(min(tsUK),max(tsUK))
fig.autofmt_xdate()
plt.grid(True)
plt.show()
#print(california_data['Population'].mean())
#print(california_data['Population'].median())
#print(california_data['Population'].mode())
```
